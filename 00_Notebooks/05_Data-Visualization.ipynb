This tutorial will walk through some basics when it comes to visualizing data with Python. Before we get started, you'll want to make sure **matplotlib** and **seaborn** are installed on your local machine (use pip).

```python
# Import modules
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

The "**%matplotlib inline**" above uses the magic operator - the "%" 
<br> <br>
That basically tells the Jupyter Notebook that matplotlib objects should be displayed within the notebook itself (rather than creating the graphic but not displaying it). Let's import some data to play with. Luckily for us, Seaborn has several native datasets that come pre-installed with the package.

```python
# Load "Tips" dataset from native Seaborn folder
data = sns.load_dataset("tips")

# Display first five rows
data.head()
```

```python
# Describe makeup of dataset
data.describe()
```

```python
# Visualize sex of servers in the present dataset
sns.countplot(x=data["sex"])
```

```python
# What if we wanted to change the dimensions of the plot?

plt.figure(figsize=(12,10))                   # Input your desired new dimensions
sns.countplot(x=data["sex"])                  # Plot the data
plt.show()                                    # Hide the matplotlib output text (only for aesthetics!)
```

```python
# What if we wanted to explore multiple variables in one plot?

plt.figure(figsize=(12,10))

# Hue == Fill value
# Palette == Fill color
sns.countplot(x=data["sex"], hue=data["time"], palette="viridis")
plt.xlabel("Sex")
plt.ylabel("")
plt.title("Count of Sex Per Time of Day")
plt.show()
```

Simple enough! For more color palette options, see [THIS LINK](https://seaborn.pydata.org/tutorial/color_palettes.html). 

<br> <br>
Next, let's visualize some numeric data to see how Seaborn handles it.

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x=data["total_bill"], y=data["tip"], s=75, alpha=0.65)
plt.xlabel("Total Bill")
plt.ylabel("Tip Left")
plt.show()
```

```python
# Add a regression line

# NOTE that the lmplot() graph takes slightly different arguments than the scatterplot() graph
sns.lmplot(data=data, x="total_bill", y="tip", aspect=2, hue="sex", height=7)
plt.xlabel("Total Bill")
plt.ylabel("Tip Left")
plt.show()
```

```python
# Seaborn also has a jointplot that combines scatterplots and histograms
sns.jointplot(x=data["total_bill"], y=data["tip"], height=10)
```

```python
# Jointplots have different "kinds" of display options
sns.jointplot(x=data["total_bill"], y=data["tip"], height=10, kind="hex")
```

```python
sns.jointplot(x=data["total_bill"], y=data["tip"], height=10, kind="kde")
```
